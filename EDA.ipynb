# EDA

The first thing to do in any project is to understand the data.

This data correspondes to historical data for bike sharing in London a dataset that is available in Kaggle ([Link](https://www.kaggle.com/datasets/hmavrodiev/london-bike-sharing-dataset)).

A couple of things are learned from the actual information regarding the dataset:s
- *timestamp* - timestamp field for grouping the data
- *cnt* - the count of a new bike shares
- *t1* - real temperature in C
- *t2* - temperature in C "feels like"
- *hum* - humidity in percentage
- *wind_speed* - wind speed in km/h
- *weather_code* - category of the weather
- *is_holiday* - boolean field - 1 holiday / 0 non holiday
- *is_weekend* - boolean field - 1 if the day is weekend
- *season* - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

- *weathe_code* category description:
1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 2 = scattered clouds / few clouds 3 = Broken clouds 4 = Cloudy 7 = Rain/ light Rain shower/ Light rain 10 = rain with thunderstorm 26 = snowfall 94 = Freezing Fog

We will now load the data and look at the actual variables present in the dataset.

### Libraries

```python
import pandas as pd
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "london_merged.csv"

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "hmavrodiev/london-bike-sharing-dataset",
    file_path
)

print(df.head())
```

```python
# Get some basic info about the data
df.info()
```

```python
# Convert the object to a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# Check for null values
print(f"There are {df.isnull().values.sum()} null values")
```

```python
df.groupby('is_holiday').count()
```
